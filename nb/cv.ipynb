```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../images/1.jpg')
```

```python
plt.imshow(img)
```

```python
img.shape
```

```python
ROI = img[40:80,30:130]
plt.imshow(ROI)
```

```python
cv2.imshow('ROI', ROI)
```

```python
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
gaussian = cv2.GaussianBlur(img2, (5, 5), 0)
plt.imshow(gaussian)
```

```python
img_p = cv2.divide(img2, gaussian, scale=255)
plt.imshow(img_p)
```

```python
cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
    if not ret:
        break
    cv2.imshow('Live Camera', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
