### unsupervised learning

```python
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
salary = random.choices(range(10000, 99999), k=1000)
age = random.choices(range(20, 60), k=1000)

df = pd.DataFrame({'salary': salary, 'age': age})
```

```python
plt.scatter(df['salary'], df['age'])
```

```python
model = KMeans(n_clusters=4)
```

```python
c = model.fit_predict(df[['salary', 'age']])
```

```python
df['status'] = c
```

```python
plt.scatter(df['salary'], df['age'], c=c)
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], s=200, c='red', marker='X')
```
